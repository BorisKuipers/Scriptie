# Resultaten visualisaties van alle  modellen

```python
import matplotlib.pyplot as plt

# === Validation Set: RMSE ===
# Sort models and RMSE values for validation set
rmse_val_sorted = sorted([(m, r) for m, r in zip(models_val, rmse_val)], key=lambda x: x[1])
models_val_rmse = [x[0] for x in rmse_val_sorted]
rmse_val_sorted = [x[1] for x in rmse_val_sorted]

# === Validation Set: MAPE ===
# Sort models and MAPE values for validation set
mape_val_sorted = sorted([(m, r) for m, r in zip(models_val, mape_val)], key=lambda x: x[1])
models_val_mape = [x[0] for x in mape_val_sorted]
mape_val_sorted = [x[1] for x in mape_val_sorted]

# === Test Set: RMSE ===
# Sort models and RMSE values for test set
rmse_test_sorted = sorted([(m, r) for m, r in zip(models_val, rmse_test)], key=lambda x: x[1])
models_test_rmse = [x[0] for x in rmse_test_sorted]
rmse_test_sorted = [x[1] for x in rmse_test_sorted]

# === Test Set: MAPE ===
# Sort models and MAPE values for test set
mape_test_sorted = sorted([(m, r) for m, r in zip(models_val, mape_test)], key=lambda x: x[1])
models_test_mape = [x[0] for x in mape_test_sorted]
mape_test_sorted = [x[1] for x in mape_test_sorted]

# === Kleuren: blauw voor feature modellen, rood voor baseline modellen ===
colors_rmse_val = ['#2E86C1' if 'Baseline' not in m else '#FF9999' for m in models_val_rmse]
colors_mape_val = ['#2E86C1' if 'Baseline' not in m else '#FF9999' for m in models_val_mape]
colors_rmse_test = ['#2E86C1' if 'Baseline' not in m else '#FF9999' for m in models_test_rmse]
colors_mape_test = ['#2E86C1' if 'Baseline' not in m else '#FF9999' for m in models_test_mape]

# === Plot: RMSE Validation Set ===
plt.figure(figsize=(12, 6))
plt.bar(models_val_rmse, rmse_val_sorted, color=colors_rmse_val)
plt.xticks(rotation=45, ha='right')
plt.ylabel('RMSE ($)', fontsize=12, fontweight='bold')
plt.title('RMSE Scores on Validation Set', pad=20, fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# === Plot: MAPE Validation Set ===
plt.figure(figsize=(12, 6))
plt.bar(models_val_mape, mape_val_sorted, color=colors_mape_val)
plt.xticks(rotation=45, ha='right')
plt.ylabel('MAPE (%)', fontsize=12, fontweight='bold')
plt.title('MAPE Scores on Validation Set', pad=20, fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# === Plot: RMSE Test Set ===
plt.figure(figsize=(12, 6))
plt.bar(models_test_rmse, rmse_test_sorted, color=colors_rmse_test)
plt.xticks(rotation=45, ha='right')
plt.ylabel('RMSE ($)', fontsize=12, fontweight='bold')
plt.title('RMSE Scores on Test Set', pad=20, fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# === Plot: MAPE Test Set ===
plt.figure(figsize=(12, 6))
plt.bar(models_test_mape, mape_test_sorted, color=colors_mape_test)
plt.xticks(rotation=45, ha='right')
plt.ylabel('MAPE (%)', fontsize=12, fontweight='bold')
plt.title('MAPE Scores on Test Set', pad=20, fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
