### Fear & Greed index data Verkennen

```python
import requests
import pandas as pd
from datetime import datetime
import time

def load_fear_greed_data(limit=0):
    """
    Load Fear & Greed Index data from the API
    Args:
        limit (int): Number of results to return. Use 0 for all available data
    Returns:
        DataFrame with Fear & Greed Index data
    """
    # API endpoint
    url = f"https://api.alternative.me/fng/?limit={limit}"
    
    try:
        # Make API request
        response = requests.get(url)
        response.raise_for_status()  
        
        # Parse JSON response
        data = response.json()
        
        # Extract the data points
        data_points = data['data']
        
        # Convert to DataFrame
        df_fear_greed = pd.DataFrame(data_points)
        
        # Convert timestamp to datetime
        df_fear_greed['date'] = pd.to_datetime(df_fear_greed['timestamp'].astype(int), unit='s')
        
        # Convert value to numeric
        df_fear_greed['value'] = df_fear_greed['value'].astype(int)
        
        # Reorder columns
        df_fear_greed = df_fear_greed[['date', 'value', 'value_classification']]
        
        # Sort by date
        df_fear_greed = df_fear_greed.sort_values('date')
        
        return df_fear_greed
        
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

if __name__ == "__main__":
    # Load all available data
    df_fear_greed = load_fear_greed_data(limit=0)
    
    if df_fear_greed is not None:
        # Save to CSV
        df_fear_greed.to_csv('fear_greed_data.csv', index=False)
        print("Data successfully saved to fear_greed_data.csv")
        print("\nFirst few rows of the data:")
        print(df_fear_greed.head())
```

```python
features_to_analyze = ['mempool_transaction_count', 'MempoolSizeBytes', 'avg_fee_usd', 'Close']

# Calculate different correlation types
pearson_corr = df_alle_features[features_to_analyze].corr(method='pearson')
spearman_corr = df_alle_features[features_to_analyze].corr(method='spearman')
kendall_corr = df_alle_features[features_to_analyze].corr(method='kendall')

# Create separate plots for each correlation type
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Pearson Correlation')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Spearman Correlation') 
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(kendall_corr, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Kendall Correlation')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Print detailed correlation analysis
print("\nDetailed Correlation Analysis:")
for feature1 in features_to_analyze:
    for feature2 in features_to_analyze:
        if feature1 < feature2:  # Avoid printing duplicate correlations
            print(f"\nCorrelation between {feature1} and {feature2}:")
            print(f"Pearson: {pearson_corr.loc[feature1, feature2]:.3f}")
            print(f"Spearman: {spearman_corr.loc[feature1, feature2]:.3f}")
            print(f"Kendall: {kendall_corr.loc[feature1, feature2]:.3f}")
```
